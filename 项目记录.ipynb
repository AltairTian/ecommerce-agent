## 电商数据分析 AI Agent
功能：

用户输入问题
↓
Agent理解需求
↓
调用SQL分析工具
↓
返回指标/图表/结论


例如：

“哪个州销售额最高？”
“近几个月GMV趋势如何？”
“延迟配送率是多少？”
“评价最低的品类有哪些？”
